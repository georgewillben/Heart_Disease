# Feature Engineering

## Objectives
* Deal with multicolinearity
* Remove any unnecessary features
* Experiment with new features

#### Import libraries and load data

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.feature_selection import chi2

# load data

train = pd.read_csv("train.csv")
val = pd.read_csv("val.csv")
explore = pd.concat([train, val])


# one hot encode catigorical columns

# get categorical columns
cat_cols = []
for col in explore.columns:
    if (explore[col].nunique() < 10) & (col != "target"):
        cat_cols.append(col)
        
# get dummies
dummies = pd.DataFrame([])
for col in cat_cols:
    dummy_col = pd.get_dummies(explore[col], prefix=col, drop_first=True)
    dummies = pd.concat([dummies, dummy_col], axis=1)

# replace columns with dummies
explore_dummies = explore.copy()
explore_dummies.drop(cat_cols, axis=1, inplace=True)
explore_dummies = pd.concat([explore_dummies, dummies], axis=1)

explore_dummies.head()
```

# 1. Check multicolinearity

```python
plt.figure(figsize=(18, 18))
sns.heatmap(explore_dummies.corr(), cmap="coolwarm", annot=True)
plt.show()
```

It seems that multicolinearity is not a problem because the highest levels of it are still below 0.60.

# 2. Remove unecessary columns

I will check to see if a column is not necessary by using the chi2 test.

```python
chi2_vals = chi2(explore_dummies, explore_dummies.target)[1]

plt.figure(figsize=(6, 5))
plt.barh(explore_dummies.columns, chi2_vals)
plt.axvline(0.05, color="orange")
plt.show()
```

It seems some features are not relevant. They are:
* fbs_1, cp_3, restecg_2, and ca_4

```python
explore_dummies.drop(["fbs_1", "cp_3", "restecg_2", "ca_4"], axis=1, inplace=True)

chi2_vals = chi2(explore_dummies, explore_dummies.target)[1]

plt.figure(figsize=(6, 5))
plt.barh(explore_dummies.columns, chi2_vals)
plt.axvline(0.05, color="orange")
plt.show()
```
